#### Identify Dissatisfied Employees

```python
dete_resignations.info()
```

```python
tafe_resignations.info()
```

```python
tafe_resignations['Contributing Factors. Job Dissatisfaction'].value_counts()
```

```python
tafe_resignations['Contributing Factors. Dissatisfaction'].value_counts()
```

```python
def update_vals(x):
    if x == '-':
        return False
    elif x == 'Not Stated':
        return np.nan
    else:
        return True
```

```python
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].map(update_vals).any(axis=1, skipna=False)
tafe_resignations_up = tafe_resignations.copy()
```

```python
tafe_resignations['dissatisfied'].value_counts()
```

```python
dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].any(axis=1, skipna=False)
dete_resignations_up = dete_resignations.copy()
dete_resignations_up['dissatisfied'].value_counts(dropna=False)
```

Above, we created a new column to identify is there were any type of jon dissatisfactions and assigned them as boolians.

#### Combine the Data

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], axis=0)
combined.info()
```

```python
combined_updated = combined.dropna(thresh=500, axis=1).copy()
combined_updated.info()
```

Above, we combined DEFE and TAFE surveys and got rid of columns that had less then 500 non-null values.

#### Clean the Service Column

```python
combined_updated['institute_service'].value_counts(dropna=False)
```

```python
combined_updated['institute_service_up'] = combined_updated['institute_service'].astype('str').str.extract(r'(\d+)')
combined_updated['institute_service_up'] = combined_updated['institute_service_up'].astype('float')
combined_updated['institute_service_up'].value_counts(dropna=False)
```

```python
def career_stage(val):
    if val >= 11:
        return "Veteran"
    elif 7 <= val < 11:
        return "Established"
    elif 3 <= val < 7:
        return "Experienced"
    elif pd.isnull(val):
        return np.nan
    else:
        return "New"
```

```python
combined_updated['work_experience'] =\
combined_updated['institute_service_up'].apply(career_stage)
combined_updated['work_experience'].value_counts(dropna=False)
```

Now we extracted precise values and categorized them.

#### Perform Initial Analysis

```python
combined_updated['dissatisfied'].value_counts(dropna=False)
```

```python
dissatisfied_pivot = combined_updated.pivot_table(index='work_experience', values='dissatisfied')

%matplotlib inline
dissatisfied_pivot.plot(kind='bar', rot=30)
plt.show()
```

Now, we can conclude that veteran and established groups are more likely to quit while experienced and newbies have not burned out yet.
